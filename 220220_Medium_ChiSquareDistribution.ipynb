```python
import numpy as np
import pandas as pd
from scipy.special import factorial
from scipy.stats import chi2
import plotly.express as px
```

```python
x_cont = np.arange(0, 15+1, 0.01, dtype='float').round(2)
```

```python
#Chi Square
lambda_list = np.array([0.2, 0.5, 1, 2, 5, 10, 15, 20], dtype='float')
pdf = pd.DataFrame(index = x_cont)
cdf = pd.DataFrame(index = x_cont)

for p in lambda_list:
  pdf[str(p)] = chi2.pdf(x_cont, p)
  cdf[str(p)] = chi2.cdf(x_cont, p)

pdf.round(2)
```

```python
cdf.round(2)
```

```python
cdf.reset_index(drop=False, inplace=True)
cdf_melt = cdf.melt(id_vars=['index']).round(5)
cdf_melt.rename(columns = {'index':'y', 'variable': 'p parameter', 'value': 'F(y)'}, inplace=True)

pdf.reset_index(drop=False, inplace=True)
pdf_melt = pdf.melt(id_vars=['index'])
pdf_melt.rename(columns = {'index':'y', 'variable': 'p parameter', 'value': 'f(y)'}, inplace=True)
```

```python
df = cdf_melt
  
fig = px.scatter(df, x="y", y="F(y)", 
                 title="Chi Square CDF",
                 animation_frame="p parameter",
                 range_y=[0,1],
                 range_x=[min(x_cont),max(x_cont)])
fig.show()
```

```python
df = pdf_melt
  
fig = px.line(df, x="y", y="f(y)", 
                 title="Chi Square PDF",
                 animation_frame="p parameter",
                 range_y=[0,1],
                 range_x=[min(x_cont), max(x_cont)])
fig.show()
```
