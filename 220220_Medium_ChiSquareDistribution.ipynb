```python
import numpy as np
import pandas as pd
from scipy.special import factorial
from scipy.stats import chi2
import plotly.express as px

from plotly.subplots import make_subplots
import plotly.graph_objects as go
!pip install -U kaleido
!wget https://github.com/plotly/orca/releases/download/v1.2.1/orca-1.2.1-x86_64.AppImage -O /usr/local/bin/orca
!chmod +x /usr/local/bin/orca
!apt-get install xvfb libgtk2.0-0 libgconf-2-4
from plotly.io import write_image
```

```python
x_cont = np.arange(0, 15+1, 0.01, dtype='float').round(2)
```

```python
#Chi Square
lambda_list = np.arange(1, 20+1, 1)
pdf = pd.DataFrame(index = x_cont)
cdf = pd.DataFrame(index = x_cont)

for p in lambda_list:
  pdf[str(p)] = chi2.pdf(x_cont, p)
  cdf[str(p)] = chi2.cdf(x_cont, p)

pdf.round(2)
```

```python
cdf.round(2)
```

```python
cdf.reset_index(drop=False, inplace=True)
cdf_melt = cdf.melt(id_vars=['index']).round(5)
cdf_melt.rename(columns = {'index':'y', 'variable': 'p parameter', 'value': 'F(y)'}, inplace=True)

pdf.reset_index(drop=False, inplace=True)
pdf_melt = pdf.melt(id_vars=['index'])
pdf_melt.rename(columns = {'index':'y', 'variable': 'p parameter', 'value': 'f(y)'}, inplace=True)
```

```python
df = cdf_melt
  
fig = px.scatter(df, x="y", y="F(y)", 
                 title="Chi Square CDF",
                 animation_frame="p parameter",
                 range_y=[0,1],
                 range_x=[min(x_cont),max(x_cont)])
fig.show()
```

```python
#subplot position list
x_dim = 4
y_dim = 5
place = []
for i in range(x_dim):
  for j in range(y_dim):
    place.append((i+1,j+1))
place
```

```python
titles = ['p = '+str(i) for i in lambda_list]
fig = make_subplots(rows=x_dim, cols=y_dim, 
                    shared_xaxes=True, shared_yaxes=True, 
                    subplot_titles=titles)

for i in range(len(lambda_list)):
  df = cdf_melt[cdf_melt['p parameter'] == str(lambda_list[i])]

  figi = px.scatter(df, x="y", y="F(y)", 
                  range_y=[0,1],
                  range_x=[min(x_cont),max(x_cont)])
  figurei_traces = []
  for trace in range(len(figi["data"])):
    figurei_traces.append(figi["data"][trace])

  for traces in figurei_traces:
      fig.add_trace(traces, row=place[i][0], col=place[i][1])
  fig.update_yaxes(range=[0, 1], row=place[i][0], col=place[i][1])

fig.update_layout(height=600, width=800, title_text="Chi Square CDFs")
fig.update_yaxes(title_text="F(y)", row=1, col=1)
fig.update_xaxes(title_text="y", row=1, col=1)

fig.write_html("Chi_Square_CDF.html")
fig.write_image("Chi_Square_CDF.png")
fig.show()
```

```python
df = pdf_melt[pd.to_numeric(pdf_melt['p parameter']) % 2 == 0] #had to limit to even parameters only - to reduce size for embed hosting
  
fig = px.line(df, x="y", y="f(y)", 
                 title="Chi Square PDF",
                 animation_frame="p parameter",
                 range_y=[0,1],
                 range_x=[min(x_cont), max(x_cont)])
fig.show()
```
