```python
import numpy as np
import pandas as pd
import math
import plotly.graph_objects as go
from scipy.special import factorial
from scipy.stats import geom
import plotly.express as px
```

```python
x_disc = np.arange(0, 15+1, 1)
x_cont = np.arange(0, 15+1, 0.01, dtype='float').round(2)
```

```python
#Geometric
lambda_list = np.arange(0.1, 1, 0.1).round(2)
pmf = pd.DataFrame(index = x_disc)
cdf = pd.DataFrame(index = x_cont)

for p in lambda_list:
  pmf[str(p)] = geom.pmf(x_disc, p)
  cdf[str(p)] = geom.cdf(x_cont, p)

pmf.round(2)
```

```python
cdf.round(2)
```

```python
cdf.reset_index(drop=False, inplace=True)
cdf_melt = cdf.melt(id_vars=['index']).round(5)
cdf_melt.rename(columns = {'index':'y', 'variable': 'p parameter', 'value': 'F(y)'}, inplace=True)

pmf.reset_index(drop=False, inplace=True)
pmf_melt = pmf.melt(id_vars=['index'])
pmf_melt.rename(columns = {'index':'y', 'variable': 'p parameter', 'value': 'P(Y=y)'}, inplace=True)
```

```python
df = cdf_melt
  
fig = px.scatter(df, x="y", y="F(y)", 
                 title="Geometric CDF",
                 animation_frame="p parameter",
                 range_y=[0,1],
                 range_x=[min(x_disc),max(x_disc)])
fig.show()
```

```python
cdf_melt[cdf_melt['p parameter'] == str(0.1)]
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
#!pip install -U kaleido
#!wget https://github.com/plotly/orca/releases/download/v1.2.1/orca-1.2.1-x86_64.AppImage -O /usr/local/bin/orca
#!chmod +x /usr/local/bin/orca
#!apt-get install xvfb libgtk2.0-0 libgconf-2-4
#import kaleido
from plotly.io import write_image

titles = ['p = '+str(i) for i in lambda_list]
fig = make_subplots(rows=3, cols=3, subplot_titles=titles)
place = [(1,1),(1,2),(1,3),(2,1),(2,2),(2,3),(3,1),(3,2),(3,3)]


for i in range(len(lambda_list)):
  df = cdf_melt[cdf_melt['p parameter'] == str(lambda_list[i])]

  figi = px.scatter(df, x="y", y="F(y)", 
                  title="Geometric CDF with p=" + str(lambda_list[i]),
                  range_y=[0,1],
                  range_x=[min(x_disc),max(x_disc)])
  figurei_traces = []
  for trace in range(len(figi["data"])):
    figurei_traces.append(figi["data"][trace])

  for traces in figurei_traces:
      fig.add_trace(traces, row=place[i][0], col=place[i][1])

fig.update_layout(height=600, width=800, title_text="Geometric CDFs")
fig.write_html("Geometric_CDF.html")
fig.write_image("Geometric_CDF.png")
fig.show()
```
