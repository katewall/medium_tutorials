```python
import numpy as np
import pandas as pd
import math
import plotly.graph_objects as go
from scipy.special import factorial
from scipy.stats import poisson
import plotly.express as px
```

```python
x_disc = np.arange(0, 15+1, 1)
x_cont = np.arange(0, 15+1, 0.01, dtype='float').round(2)
```

```python
#Poisson
lambda_list = np.array([0.2, 0.5, 1, 2, 5, 10, 15, 20], dtype='float')
pois_pmf = pd.DataFrame(index = x_disc)
pois_cdf = pd.DataFrame(index = x_cont)

for p in lambda_list:
  pois_pmf[str(p)] = poisson.pmf(x_disc, p)
  pois_cdf[str(p)] = poisson.cdf(x_cont, p)

pois_pmf.round(2)
```

```python
pois_cdf.round(3)
```

```python
pois_cdf.reset_index(drop=False, inplace=True)
pois_cdf_melt = pois_cdf.melt(id_vars=['index']).round(5)
pois_cdf_melt.rename(columns = {'index':'y', 'variable': 'lambda parameter', 'value': 'F(y)'}, inplace=True)

pois_pmf.reset_index(drop=False, inplace=True)
pois_pmf_melt = pois_pmf.melt(id_vars=['index'])
pois_pmf_melt.rename(columns = {'index':'y', 'variable': 'lambda parameter', 'value': 'P(Y=y)'}, inplace=True)
```

```python
df = pois_cdf_melt
  
fig = px.scatter(df, x="y", y="F(y)", 
                 title="Poisson CDF",
                 animation_frame="lambda parameter",
                 range_y=[0,1])

fig2 = go.Figure() #new figure for resetting first frame
for tr in fig.frames[-4].data: # add last frame traces to fig2
    fig2.add_trace(tr)

fig2.layout = fig.layout #copy layout
fig2.frames = fig.frames #copy frames
fig2.layout['sliders'][0]['active'] = len(fig.frames) - 4 # set 4th to last frame as the active one
fig2.show()
```

```python
df = pois_pmf_melt
  
fig = px.bar(df, x="y", y="P(Y=y)", 
                 title="Poisson PMF",
                 animation_frame="lambda parameter",
                 range_y=[0,0.6],
                 range_x=[min(x_disc), max(x_disc)])
#fig["layout"].pop("updatemenus") #good for toggle only, no play button

fig2 = go.Figure() #new figure for resetting first frame
for tr in fig.frames[-4].data: # add last frame traces to fig2
    fig2.add_trace(tr)

fig2.layout = fig.layout #copy layout
fig2.frames = fig.frames #copy frames
fig2.layout['sliders'][0]['active'] = len(fig.frames) - 4 # set 4th to last frame as the active one

fig2.show()
```
