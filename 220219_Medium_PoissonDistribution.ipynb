```python
import numpy as np
import pandas as pd
import math
import plotly.graph_objects as go
from scipy.special import factorial
from scipy.stats import poisson
import plotly.express as px
```

```python
x_disc = np.arange(-2, 5+1, 1)
x_cont = np.arange(-2, 5+1, 0.01, dtype='float').round(2)
```

```python
#Poisson
lambda_list = np.array([0.2, 0.5, 1, 2, 7, 12], dtype='float')
pois_pmf = pd.DataFrame(index = x_disc)
pois_cdf = pd.DataFrame(index = x_cont)

for p in lambda_list:
  pois_pmf[str(p)] = poisson.pmf(x_disc, p)
  pois_cdf[str(p)] = poisson.cdf(x_cont, p)

pois_pmf.round(2)
```

```python
pois_cdf.round(3)
```

```python
pois_cdf.reset_index(drop=False, inplace=True)
pois_cdf_melt = pois_cdf.melt(id_vars=['index']).round(5)
pois_cdf_melt.rename(columns = {'index':'y', 'variable': 'lambda parameter', 'value': 'CDF'}, inplace=True)

pois_pmf.reset_index(drop=False, inplace=True)
pois_pmf_melt = pois_pmf.melt(id_vars=['index'])
pois_pmf_melt.rename(columns = {'index':'y', 'variable': 'lambda parameter', 'value': 'PMF'}, inplace=True)
```

```python
df = pois_cdf_melt
  
fig = px.scatter(df, x="y", y="CDF", 
                 title="Poisson CDF of y",
                 animation_frame="lambda parameter")
  
fig["layout"].pop("updatemenus")
fig.show()
```

```python
df = pois_pmf_melt
  
fig = px.bar(df, x="y", y="PMF", 
                 title="Poisson PMF",
                 animation_frame="lambda parameter")
  
fig["layout"].pop("updatemenus")
fig.show()
```
