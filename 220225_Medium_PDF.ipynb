```python
from collections import Counter
from matplotlib_venn import venn3

sets = Counter()               #set order A, B, C  
sets['100'] = 4      #100 denotes A on, B off, C off
sets['001'] = 1      #001 denotes A off, B off, C on
sets['010'] = 3      #010 denotes A off, B on, C off
sets['101'] = 5     #101 denotes A on, B off, C on
sets['110'] = 13     #110 denotes A on, B on, C off
sets['011'] = 15     #011 denotes A off, B on, C on
sets['111'] = 20 #011 denotes A on, B on, C on
labels = ('Group A', 'Group B', 'Group C')

plt.figure(figsize=(5,5))
ax = plt.gca()
v = venn3(subsets=sets, set_labels=('','',''), ax=ax,set_colors=('darkviolet','deepskyblue','blue'),alpha=0.7)
for text in v.subset_labels:
  text.set_color('white')
  text.set_fontsize(12)
  text.set_fontweight('bold')


h = [] 
for i in sets:
    h.append(v.get_patch_by_id(i)) #get color patches 
l = ['A only','C only','B only','A & C shared','A & B shared','B & C shared','A & B & C shared']
ax.legend(handles=h[0:3], labels=l[0:3], title="Legend",loc='lower left') #create tailored legend from handles and labels    
plt.title('Membership across Three Groups')
plt.show()
```

```python
import pandas as pd
import matplotlib as mpl
deepgreen = '#8c9344'
lightbeige = '#eef1d4'
medgreen = '#a2a963'
darkbrown = '#242112'
niceorange = '#Ffa800'


fig = plt.figure(figsize=(8.5, 11), facecolor=deepgreen)
rows = 3
cols = 2
grid = plt.GridSpec(rows, cols, wspace=0.2, hspace=0.15)

#Plot
col = 'darkolivegreen'
col = darkbrown
fig_ax1 = fig.add_subplot(grid[0, 0], facecolor=medgreen)# title='Title')
fig_ax1.spines['bottom'].set_color(col)
fig_ax1.spines['top'].set_color(col) 
fig_ax1.spines['right'].set_color(col)
fig_ax1.spines['left'].set_color(col)
fig_ax1.tick_params(axis='x', colors=col)#ticks
fig_ax1.tick_params(axis='y', colors=col)
fig_ax1.yaxis.label.set_color(col)       #axis label
fig_ax1.xaxis.label.set_color(col)       
fig_ax1.title.set_color(col)             #title
fig_ax1 = plt.plot(range(7), [3, 1, 4, 1, 5, 0, 2], 'w-o')

desc1 = 'A short description here \n line break continues \n the interesting information.'
desc2 = '\n \n We can fill the space \n so that it looks nicely placed.'
desc3 = '\n \n ~ bullet 1 \n ~ bullet 2 \n ~ bullet 3'
fig.text(1/cols+0.025, 2/rows+0.03, desc1+desc2+desc3, color=darkbrown) # horizontally 1/2 halfway from left, vertically 2/3 from bottom, 0,0 is bottom left, 1,1 is top right

#Table
mpl.rcParams['text.color'] = lightbeige #table text color
fig_ax3 = fig.add_subplot(grid[1, :2])
fig_ax3.set_axis_off()                  #turn off ticks
ncol = 4                                #table columns
nrow = 9                                #table rows
df = pd.DataFrame(np.random.randn(nrow, ncol).round(2), 
                  columns=['Business Size', 'Type', 'Personnel', 'Amount Raised'])
fig_ax3.table(cellText=df.values, colLabels=df.columns, 
              loc='center',             #header cell alignment
              cellLoc='center',         #body cell alignment
              cellColours=np.full((nrow, ncol), medgreen), #body cell fill color
              colColours=['olivedrab']*ncol,#header cell fill color
              edges='horizontal')       #turns edges on horizontally


#Venn
fig_ax4 = fig.add_subplot(grid[2, 1], title='Services across Three Businesses')
v = venn3(subsets=sets, set_labels=('','',''), 
          ax=fig_ax4, 
          set_colors=('darkviolet','olivedrab',niceorange),
          alpha=0.7)

for text in v.subset_labels:
  text.set_color('white')
  text.set_fontsize(12)
  text.set_fontweight('bold')

mpl.rcParams['text.color'] = darkbrown
h = [] 
for i in sets:
    h.append(v.get_patch_by_id(i)) #get color patches 
l = ['A only','C only','B only','A & C shared','A & B shared','B & C shared','A & B & C shared']
fig_ax4.legend(handles=h[0:3], labels=l[0:3], title="Legend",loc='lower left') #create tailored legend from handles and labels    

desc21 = 'More short description of how much \n time this could save us \n when communicating routine results.'
desc22 = '\n \n We can fill the space \n so that it looks EXCEPTIONALLY placed.'
fig.text(1/cols-0.055, 0/rows+0.23, desc21+desc22, color=darkbrown, horizontalalignment='right') 

fig.suptitle('One Page Brochure', fontsize=38, color=lightbeige, family='monospace')
desc3 = 'Example by Kate Wall'
fig.text(0.5, 0.91, desc3, color=lightbeige, horizontalalignment='center', family='monospace') 
plt.show()
```
